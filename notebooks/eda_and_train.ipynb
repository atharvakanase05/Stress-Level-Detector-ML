```python
# ==========================================
# Stress Detection EDA & Dataset Inspection
# ==========================================

import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 1. Set working directory
# -----------------------------
# Change to your project root (where 'data' folder exists)
os.chdir(r"C:\Users\Admin\Desktop\ML Mini Project Sem7\stress-detection")

# Verify current working directory
print("Current working directory:", os.getcwd())

# List files in project root
print("Files here:", os.listdir())

# List files in data folder
print("Files in data folder:", os.listdir("data"))

# -----------------------------
# 2. Load the dataset
# -----------------------------
dataset_path = r"data/stress_detection_data.csv"
df = pd.read_csv(dataset_path)

# -----------------------------
# 3. Preview dataset
# -----------------------------
print("\n--- First 5 rows ---")
print(df.head())

print("\n--- Dataset Summary ---")
print(df.describe(include="all"))

# -----------------------------
# 4. Target column distribution
# -----------------------------
print("\n--- Stress Level Counts ---")
print(df['Stress_Detection'].value_counts())

sns.countplot(data=df, x='Stress_Detection')
plt.title("Class Distribution of Stress Levels")
plt.show()

# -----------------------------
# 5. Correlation heatmap (numeric features only)
# -----------------------------
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()

# -----------------------------
# 6. Pairplot (small sample for speed)
# -----------------------------
sampled_df = df.sample(200, random_state=42)  # smaller sample

# Select numeric columns only
numeric_cols = sampled_df.select_dtypes(include='number').columns

# Plot pairplot
sns.pairplot(sampled_df, vars=numeric_cols, hue='Stress_Detection')
plt.show()
```
